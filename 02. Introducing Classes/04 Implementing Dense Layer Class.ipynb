```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = (0.01) * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def foreward_pass(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

In machine learning, X represents the input feature matrix (multiple samples and features), while y represents the output labels or targets (one value per sample). Uppercase X denotes a matrix, and lowercase y denotes a vector ‚Äî this follows standard mathematical and ML notation.

```python
X, y = spiral_data(samples = 100, classes = 3)

# Create Dense layer with 2 input features and 3 neurons (output values)
dense1 = Layer_Dense(2, 3)
# Perform a forward pass of our training data through this layer
dense1.foreward_pass(X)

print(dense1.output[:5])
# print(np.array(dense1.output[:5]).T)
```

```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data

# Initialize NNFS (sets random seed, dtype, etc.)
# nnfs.init()

# Dense layer
class Layer_Dense:
    # Layer initialization
    def __init__(self, n_inputs, n_neurons):
        # Initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    # Forward pass
    def forward(self, inputs):
        # Calculate output values from inputs, weights and biases
        self.output = np.dot(inputs, self.weights) + self.biases


# Create dataset
X, y = spiral_data(samples=100, classes=3)

# Create Dense layer with 2 input features and 3 output values
dense1 = Layer_Dense(2, 3)

# Perform a forward pass of our training data through this layer
dense1.forward(X)

# Display first 5 outputs
print(dense1.output[:5])
```
